#1. Análise do faturamento


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns  

# URLs dos dados das lojas
urls = [
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"
]

# Função para calcular o faturamento total
def calcular_faturamento_total(df):
    return df['Preço'].sum()

# Função para formatar valores em moeda brasileira (R$)
def formatar_moeda(valor):
    return f"R$ {valor:.2f}"

# Lendo os dados e calculando os faturamentos
faturamentos = {}
for i, url in enumerate(urls, start=1):
    df = pd.read_csv(url)
    faturamento = calcular_faturamento_total(df)
    faturamentos[f"Loja {i}"] = faturamento

# Calculando o faturamento total
faturamento_total = sum(faturamentos.values())

# Exibindo os resultados
print(f"Faturamento total: {formatar_moeda(faturamento_total)}")
for loja, faturamento in faturamentos.items():
    print(f"Faturamento {loja}: {formatar_moeda(faturamento)}")

# Visualizando o faturamento total por loja
plt.figure(figsize=(10, 6))
sns.barplot(x=list(faturamentos.keys()), y=list(faturamentos.values()), palette="Blues")

# Adicionando título e labels
plt.title('Faturamento Total por Loja', fontsize=16, pad=20)
plt.xlabel('Lojas', fontsize=12, labelpad=10)
plt.ylabel('Faturamento Total (R$)', fontsize=12, labelpad=10)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

# Exibindo o gráfico
plt.tight_layout()
plt.show()

# Faturamento total por loja e por produto
for i, url in enumerate(urls, start=1):
    df = pd.read_csv(url)
    faturamento_por_produto = df.groupby("Produto")["Preço"].sum().reset_index()
    faturamento_por_produto = faturamento_por_produto.sort_values(by="Preço", ascending=False).head(10)  # Top 10 produtos
    print(f"\nFaturamento por produto na Loja {i}:")
    print(faturamento_por_produto)

    # Visualizando o faturamento por produto
    plt.figure(figsize=(12, 6))
    sns.barplot(x="Produto", y="Preço", data=faturamento_por_produto, palette="Greens")

    # Adicionando título e labels
    plt.title(f'Top 10 Produtos - Faturamento na Loja {i}', fontsize=16, pad=20)
    plt.xlabel('Produto', fontsize=12, labelpad=10)
    plt.ylabel('Faturamento (R$)', fontsize=12, labelpad=10)
    plt.xticks(rotation=45, fontsize=10, ha='right')
    plt.yticks(fontsize=10)

    # Exibindo o gráfico
    plt.tight_layout()
    plt.show()
```

# 3. Média de Avaliação das Lojas

```python
import pandas as pd
import matplotlib.pyplot as plt

# URLs dos dados das lojas
urls = [
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"
]

# Função para calcular a média das avaliações
def calcular_media_avaliacoes(df):
    return df['Avaliação da compra'].mean()

# Lendo os dados e calculando as médias
medias = {}
for i, url in enumerate(urls, start=1):
    df = pd.read_csv(url)
    media = calcular_media_avaliacoes(df)
    medias[f"Loja {i}"] = media
    print(f"Média de avaliações Loja {i}: {media:.2f}")

# Preparando os dados para o gráfico
lojas = list(medias.keys())
media_avaliacoes = list(medias.values())

# Criando o gráfico de barras
plt.bar(lojas, media_avaliacoes, color='skyblue')

# Adicionando rótulos nas barras
for i, media in enumerate(media_avaliacoes):
    plt.text(i, media + 0.1, f"{media:.2f}", ha='center')

# Personalizando o gráfico
plt.title('Média das Avaliações por Loja', fontsize=16)
plt.xlabel('Lojas', fontsize=12)
plt.ylabel('Média das Avaliações', fontsize=12)
plt.ylim(0, 5)  # Definindo o limite do eixo y para 0 a 5
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Exibindo o gráfico
plt.show()
```

# 4. Produtos Mais e Menos Vendidos

```python
import pandas as pd
import matplotlib.pyplot as plt

# URLs dos dados das lojas
urls = [
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"
]

# Função para identificar os produtos mais e menos vendidos
def identificar_produtos_mais_e_menos_vendidos(df):
    vendas_por_produto = df.groupby("Produto")["Quantidade de parcelas"].sum().reset_index()
    vendas_por_produto.rename(columns={"Quantidade de parcelas": "Quantidade"}, inplace=True)
    mais_vendidos = vendas_por_produto.nlargest(5, "Quantidade")
    menos_vendidos = vendas_por_produto.nsmallest(5, "Quantidade")
    return mais_vendidos, menos_vendidos

# Lendo os dados e identificando os produtos mais e menos vendidos
resultados = {}
for i, url in enumerate(urls, start=1):
    df = pd.read_csv(url)
    mais_vendidos, menos_vendidos = identificar_produtos_mais_e_menos_vendidos(df)
    resultados[f"Loja {i}"] = {
        "Mais Vendidos": mais_vendidos,
        "Menos Vendidos": menos_vendidos
    }

# Exibindo os resultados
for loja, dados in resultados.items():
    print(f"\n{loja}:")
    print("Produtos mais vendidos:")
    print(dados["Mais Vendidos"])
    print("\nProdutos menos vendidos:")
    print(dados["Menos Vendidos"])

# Visualizando os resultados em gráficos
for loja, dados in resultados.items():
    plt.figure(figsize=(12, 6))
    
    # Gráfico dos produtos mais vendidos
    plt.subplot(1, 2, 1)
    plt.bar(dados["Mais Vendidos"]["Produto"], dados["Mais Vendidos"]["Quantidade"], color='green')
    plt.title(f'{loja} - Produtos Mais Vendidos')
    plt.xlabel('Produto')
    plt.ylabel('Quantidade Vendida')
    plt.xticks(rotation=45)
    
    # Gráfico dos produtos menos vendidos
    plt.subplot(1, 2, 2)
    plt.bar(dados["Menos Vendidos"]["Produto"], dados["Menos Vendidos"]["Quantidade"], color='red')
    plt.title(f'{loja} - Produtos Menos Vendidos')
    plt.xlabel('Produto')
    plt.ylabel('Quantidade Vendida')
    plt.xticks(rotation=45)
    
    plt.tight_layout()
    plt.show()
```

# 5. Média de frete por Loja

```python
import pandas as pd
import matplotlib.pyplot as plt

# URLs dos dados das lojas
urls = [
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"
]

# Função para calcular o custo médio de frete
def calcular_custo_medio_frete(df):
    return df['Frete'].mean()

# Lendo os dados e calculando os custos médios
custo_medio_frete = {}
for i, url in enumerate(urls, start=1):
    df = pd.read_csv(url)
    custo_medio = calcular_custo_medio_frete(df)
    custo_medio_frete[f"Loja {i}"] = custo_medio
    print(f"Custo médio de frete Loja {i}: {custo_medio:.2f}")

# Preparando os dados para o gráfico
lojas = list(custo_medio_frete.keys())
custo_frete = list(custo_medio_frete.values())

# Criando o gráfico de barras
plt.figure(figsize=(10, 6))
bars = plt.bar(lojas, custo_frete, color='orange')

# Adicionando rótulos nas barras
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, f"{height:.2f}", ha='center', va='bottom')


plt.title('Custo Médio de Frete por Loja', fontsize=16)
plt.xlabel('Lojas', fontsize=12)
plt.ylabel('Custo Médio de Frete (R$)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
